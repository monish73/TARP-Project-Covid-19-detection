```python
import pandas as pd
import os
import shutil
```

## **Data-Processing for Positive Cases**

```python
#create the data for positive sample

FILE_PATH = "chest_xray_positive/metadata.csv"
IMAGES_PATH = "chest_xray_positive/images"
```

```python
df = pd.read_csv(FILE_PATH)
print(df.shape)
```

```python
df.head()
```

```python
TARGET_DIR = "Dataset/Covid"

if not os.path.exists(TARGET_DIR):
    os.mkdir(TARGET_DIR)
    print("Covid folder created")
```

```python
cnt = 0

for(i, row) in df.iterrows():
    if row["finding"] == "Pneumonia/Viral/COVID-19":
        cnt += 1
        
print(cnt)
```

### Filter Front-view X-RAY images not other top or side views

```python
cnt = 0

for(i, row) in df.iterrows() :
    if row["finding"] == "Pneumonia/Viral/COVID-19" and row["view"] == "PA":
        cnt += 1
        
print(cnt)
```

### Copying to DATASET folder

```python
cnt = 0

for(i, row) in df.iterrows() :
    if row["finding"] == "Pneumonia/Viral/COVID-19" and row["view"] == "PA":
        filename = row["filename"]
        image_path = os.path.join(IMAGES_PATH, filename)
        image_copy_path = os.path.join(TARGET_DIR, filename)
        shutil.copy2(image_path, image_copy_path)
        print("processing image ", cnt)
        cnt += 1
        
print(cnt)
```
